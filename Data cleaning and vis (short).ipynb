```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv("activity_botscore.csv", low_memory=False)
df.head()
```

## Data organization and viz
It's already pretty clean, we can do minimal cleaning and of visualization here.

```python
df.dtypes
```

```python
df.shape
```

```python
df = df.rename(columns={'age':'Age (days)', 'count':'#ofPosts', 'activity':'Posts/Day', 'bot_score_english':'OddsBeingBot'})
df.head()
```

```python
df = df.drop('user_id', axis=1)
df.head()
```

```python
import seaborn as sns
sns.regplot(df['Posts/Day'],df['OddsBeingBot'])
```

```python
sns.regplot(df['Posts/Day'],df['OddsBeingBot'])
plt.title("Odds of being a bot by posts per day")
plt.xlabel("Posts per day")
plt.ylabel("Odds of Being a Bot")
```

```python
sns.regplot(df['Posts/Day'],df['OddsBeingBot'])
plt.title("Odds of being a bot by net posts")
plt.xlabel("Total # of Posts")
plt.ylabel("Odds of Being a Bot")
```

```python
sns.regplot(df['Age (days)'],df['OddsBeingBot'])
plt.title("Odds of being a bot by net posts")
plt.xlabel("Account Age in Days")
plt.ylabel("Odds of Being a Bot")
```
