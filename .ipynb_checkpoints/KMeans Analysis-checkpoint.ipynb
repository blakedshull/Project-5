```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv("updated.csv", low_memory=False)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

## Data Analysis

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

### KMeans

#### Inertia plot

```python
inertias = []
for k in range(2,10):
    kmeans = KMeans(n_clusters=k)  
    kmeans.fit(scaled_df)  
    inertias.append(kmeans.inertia_)
plt.plot(range(2,10), inertias, marker = '.')
plt.xlabel('clusters')
plt.ylabel('inertia')
```

Maybe four or five clusters will help me track their model

#### Silhouette scores next

```python
silhouette_scores = []
for i in range(2, 10):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(scaled_df)
  silhouette_scores.append(silhouette_score(scaled_df, kmeans.labels_))
plt.plot(range(2, 10), silhouette_scores, marker = '.')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score');
```

2 silhouette scores is optimal

### Clustering based on plots

#### Clustering based off inertia

```python
kmeans = KMeans(n_clusters = 4, random_state=42)
kmeans.fit(scaled_df)
df['cluster'] = kmeans.labels_
df.head()
```

```python
kmeans2 = KMeans(n_clusters = 5, random_state=42)
kmeans2.fit(scaled_df)
df['cluster2'] = kmeans2.labels_
df.head()
```

```python
df['cluster'] = kmeans.labels_
df['cluster2'] = kmeans2.labels_

fig, axes = plt.subplots(2,1, figsize=(5,8))
axes[0].scatter(df['OddsBeingBot'], df['#ofPosts'], c = df['cluster'])
axes[0].set_xlabel('OddsBeingBot')
axes[0].set_ylabel('#ofPosts')
axes[0].set_title('KMeans Clusters')
axes[1].scatter(df['OddsBeingBot'], df['#ofPosts'], c = df['cluster2'])
axes[1].set_xlabel('OddsBeingBot')
axes[1].set_ylabel('#ofPosts')
axes[1].set_title('KMeans 2 Clusters')
plt.tight_layout();
```

Either of these are decently accurate 

It seems like we do have a main cluster of accounts that have an odds of being a bot over 40%.  Decent match of the model, 

#### Silhouette clustering

```python
kmeans3 = KMeans(n_clusters = 2, random_state=42)
kmeans3.fit(scaled_df)
df['cluster3'] = kmeans3.labels_
df.head()
```

```python
kmeans4 = KMeans(n_clusters = 3, random_state=42)
kmeans4.fit(scaled_df)
df['cluster4'] = kmeans4.labels_
df.head()
```

```python
df['cluster3'] = kmeans3.labels_
fig, axes = plt.subplots(2,1, figsize=(5,8))
axes[0].scatter(df['OddsBeingBot'], df['#ofPosts'], c = df['cluster3'])
axes[0].set_xlabel('OddsBeingBot')
axes[0].set_ylabel('#ofPosts')
axes[0].set_title('KMeans 3 Clusters')
axes[1].scatter(df['OddsBeingBot'], df['#ofPosts'], c = df['cluster4'])
axes[1].set_xlabel('OddsBeingBot')
axes[1].set_ylabel('#ofPosts')
axes[1].set_title('KMeans 4 Clusters')
```
